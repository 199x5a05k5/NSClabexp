```python
#exp 3 caesar cipher 
c=[chr(i) for i in range(65,91)]
s=[chr(i) for i in range(97,123)]
msg=input("Enter Message:")
key=int(input("Enter shift key:"))

cipher=[]
decrypt=[]
for i in msg:
    if i in c:
        pos=(c.index(i)+key)%26
        cipher.append(c[pos])
    else:
        pos=(s.index(i)+key)%26
        cipher.append(s[pos])
print("Encrypted Message is :","".join(cipher))

cipher="".join(cipher)
for i in cipher:
    if i in c:
        pos=(c.index(i)-key)%26
        decrypt.append(c[pos])
    else:
        pos=(s.index(i)-key)%26
        decrypt.append(s[pos])
print("Decrypted message is :","".join(decrypt))
```

```python
#exp 5 Transposition chipher (Rail Fence)
print("\t\t RAIL FENCE TECHNIQUE")
print("\t\t Enter the input string")
m=input("\t ")
l=len(m)
j=0
c=[]
d=[" " for i in range(l)]
for i in range(0,l):
    if i%2==0:
        c.append(m[i])
for i in range(0,l):
    if i%2==1:
        c.append(m[i])
print("\t Cipher text after applying rail fence :","".join(c))
print("\t Deciphering")
for i in range(0,l):
    if i%2==0:
        d[i]=c[j]
        j+=1
for i in range(0,l):
    if i%2==1:
        d[i]=c[j]
        j+=1
print("\t Text after decryptioin :","".join(d))

        
```

```python
#exp 7 rsa program
def gcd(a,b):
    if a==0:
        return b
    return gcd(b%a,a)

p,q=3,7
n=p*q
phi=(p-1)*(q-1)
e=2

while(e<phi):
    count=gcd(e,phi)
    if count==1:
        break
    else:
        e+=1
#choosing d such that it satisfies d*e = 1 + k * totient
k=2
d=(1+(k*phi))/e
#s=[chr(i) for i in range(97,123)]
msg=int(input("Enter Message: "))
cipher=(msg**e)%n
decrypt=(cipher**d)%n

print("Entered original message is :",msg)
print("p=",p)
print("q=",q)
print("n=p*q",n)
print("phi(n)=(p-1)*(q-1)",phi)
print("e and d is :",e,d)
print("cipher text is :",cipher)
print("Decrypt message is :",decrypt)
```

```python
#exp 8 Diffie-Hellman Key Exchange
def power(a,b,mod):
    if b==1:
        return a
    t=power(a,b//2,mod)
    if b%2==0:
        return (t*t)%mod
    else:
        return (((t*t)%mod)*a)%mod

n,g = map(int, input("enter the value of n and g: ").split())
x=int(input("enter the value of x for first person: "))
a=power(g,x,n)
y=int(input("enter the value of y for second person: "))
b=power(g,y,n)
print("key for the first person is :",power(b,x,n))
print("key for the second person is :",power(a,y,n))
#n,g,x,y=23,9,4,3 o/p=9,9
```

```python
import random  
from math import pow
  
a = random.randint(2, 10) 
  
def gcd(a, b): 
    if a < b: 
        return gcd(b, a) 
    elif a % b == 0: 
        return b; 
    else: 
        return gcd(b, a % b) 
  
# Generating large random numbers 
def gen_key(q): 
    key = random.randint(pow(10, 20), q) 
    while gcd(q, key) != 1: 
        key = random.randint(pow(10, 20), q) 
    return key 
  
# Modular exponentiation 
def power(a, b, c): 
    x = 1
    y = a 
  
    while b > 0: 
        if b % 2 == 0: 
            x = (x * y) % c; 
        y = (y * y) % c 
        b = int(b / 2) 
  
    return x % c 
  
# Asymmetric encryption 
def encrypt(msg, q, h, g): 
  
    en_msg = [0]*len(msg)
  
    k = gen_key(q)# Private key for sender 
    s = power(h, k, q) 
    p = power(g, k, q) 
      
#     for i in range(0, len(msg)): 
#         en_msg.append(msg[i]) 
  
    print("g^k used : ", p) 
    print("g^ak used : ", s) 
    for i in range(0, len(msg)): 
        en_msg[i] = s * ord(msg[i]) 
  
    return en_msg, p 
  
def decrypt(en_msg, p, key, q): 
  
    dr_msg = [] 
    h = power(p, key, q) 
    for i in range(0, len(en_msg)): 
        dr_msg.append(chr(int(en_msg[i]/h))) 
          
    return dr_msg 
  
# Driver code 
def main(): 
  
    msg = 'encryption'
    print("Original Message :", msg) 
  
    q = random.randint(pow(10, 20), pow(10, 50)) 
    g = random.randint(2, q) 
  
    key = gen_key(q)# Private key for receiver 
    h = power(g, key, q) 
    print("g used : ", g) 
    print("g^a used : ", h) 
  
    en_msg, p = encrypt(msg, q, h, g) 
    dr_msg = decrypt(en_msg, p, key, q) 
    dmsg = ''.join(dr_msg) 
    print("Decrypted Message :", dmsg); 
  
  
if __name__ == '__main__': 
    main()
```
